# Introduction

This notebook shows how to use the forward model (GVS) that simulates light propagation in the holographic display.

<center><img alt="Experimental Setup" src="../data/experimental-setup.png" width="512"/></center>

```python
import numpy as np
import plotly.express as px
import torch

from imageio.v3 import imread

from holographic_display.forward_model import propagate
from holographic_display.constants import nm, um, mm
```

# Input images

```python
source = imread("../data/source.png") / 255
phase = imread("../data/slm.png") * 2.0 * np.pi / 255
```

```python
px.imshow(source, title="Source intensity image displayed on the iphone")
```

```python
px.imshow(phase, title="Phase image displayed on the SLM", color_continuous_scale="gray")
```

# Physical and numerical parameters

```python
nx_camera = ny_camera = 1024
dx_source = 55.2 * um
dx_slm = 3.74 * um

red = 620 * nm
green = 532 * nm
blue = 461 * nm
reference_wavelength = 460 * nm
wavelengths = [red, green, blue]

z_source_slm = 250 * mm
z_object_camera_1 = 80 * mm
z_object_camera_2 = 100 * mm
propagation_distances = [z_object_camera_1, z_object_camera_2]
```

# Results after light propagation

```python
device = "mps"  # Use cuda or cpu depending on installation
torch.set_default_device(device)
```

```python
results = propagate(
    torch.tensor(source, dtype=torch.float32),
    torch.tensor(phase, dtype=torch.float32),
    wavelengths,
    reference_wavelength,
    dx_source,
    dx_slm,
    z_source_slm,
    propagation_distances,
    nx_camera,
    ny_camera,
)
```

```python
px.imshow(results[0].cpu(), title="Result at 8 cm", origin="lower")
```

```python
px.imshow(results[1].cpu(), title="Result at 10 cm", origin="lower")
```
